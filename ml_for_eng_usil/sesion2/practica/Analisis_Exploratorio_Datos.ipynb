```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ruta del archivo CSV
file_path = 'wdbc.data'

# Leer el archivo CSV
cancer_df = pd.read_csv(file_path, header=None)

# Eliminar la primera columna (ID) y Cambiar M y B a 1 y 0 respectivamente
cancer_df = cancer_df.drop(columns=[0])
cancer_df[1] = cancer_df[1].map({'M': 1, 'B': 0})

# Ver las primeras filas del conjunto de datos
cancer_df.head()
```

# Descripción de datos

```python
# Estadísticas descriptivas
print(cancer_df.describe())
```

# Correlación

```python
# Crear una submatriz de correlación
correlation_matrix = cancer_df.iloc[:,1:].corr()

# Crear un mapa de calor de correlación
plt.figure(figsize=(50, 50))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Matriz de Correlación entre Variables Numéricas de cancer_df')
plt.show()
```

# Diagrama de Cajas

```python
# Crear el box-plot de la variable "edad"
plt.figure(figsize=(10, 6))
sns.boxplot(x=cancer_df.iloc[:,2])
plt.title('Box-Plot de la Edad')
plt.xlabel('Edad')
plt.show()
```
