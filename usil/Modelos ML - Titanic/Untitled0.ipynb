```python
def calculate_accuracy(y_true, y_pred):
  """
  Calculates the accuracy of predictions.

  Args:
    y_true: True labels.
    y_pred: Predicted labels.

  Returns:
    The accuracy as a float.
  """
  correct_predictions = (y_true == y_pred).sum()
  accuracy = correct_predictions / len(y_true)
  return accuracy
```
